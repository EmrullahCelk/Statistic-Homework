```python
# 2) Which are the two columns with the highest correlation in Titanic data?

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")
data_titanic = pd.read_csv("titanic.csv")
f,ax=plt.subplots(figsize = (10,10))
sns.heatmap(data_titanic.corr(),annot= True,linewidths=0.5,fmt = ".3f",ax=ax)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Correlation')
plt.show()
```

```python
# q3 - For Fare columns in Titanic data find
#     a) maximum, minimum
#     b) mean
#     c) mode
#     d) median and
#     f) Draw the graph boxplot

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")
data_titanic = pd.read_csv("titanic.csv", usecols=['Fare'])
data_titanic.head(n=1000)
```

```python
print("Titanic Fare Colum   :")
print('Min Value            :', np.min(data_titanic.head(n=1000)))
print('Max Value            :', np.max(data_titanic.head(n=1000)))
print('Mean Value           :', np.mean(data_titanic.head(n=1000)))
print('Mode Value           :', stats.mode(data_titanic.head(n=1000)))
print('Median Value         :', np.median(data_titanic.head(n=1000)))
```

```python
df = pd.read_csv("titanic.csv")
df.head(n=1000)
df.boxplot(column =['Fare'], grid = True)
```

```python
# 5) The column "bill_length_mm" of penguins data, find:
    # a) mean
    # b) minimum, maximum
    # c) mode
    # d) median
    # e) Draw the graph boxplot




import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")
data_penguins = pd.read_csv("penguins2.csv", usecols=['bill_length_mm'])
data_penguins.head(n=1000)
```
